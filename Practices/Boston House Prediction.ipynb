```python
house_price_dataset = sklearn.datasets.load_boston()
```

```python
print(house_price_dataset)
```

```python
#loading dataset to a pandas Dataframe
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

```python
#print 5 rows of data
house_price_dataframe.head()
```

```python
#add the taregt price column to the dataframe
house_price_dataframe['price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
#checking shape of the data(rows and Columns)
house_price_dataframe.shape
```

```python
#check for missing values
house_price_dataframe.isnull().sum()
```

```python
#statistical measures of a datset
house_price_dataframe.describe()
```

```python
from numpy.core.numeric import correlate
#1.positive correlation
#2.negative correlation
 
correlation = house_price_dataframe.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square = True ,fmt='.1f', annot_kws={'size':8},cmap='Blues')
```

```python
X = house_price_dataframe.drop(['price'], axis=1)
Y = house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.2, random_state= 2)
```

```python
print(X.shape, X_train.shape , X_test.shape)
```

```python
from xgboost.sklearn import XGBRFRegressor
model = XGBRFRegressor()
```

```python
model.fit(X_train, Y_train)
```

```python
training_data_prediction = model.predict(X_train)
```

```python
rse = metrics.r2_score(Y_train, training_data_prediction)
mae = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error: ",rse)
print("Mean absolute error: ",mae)
```

```python
test_data_prediction = model.predict(X_test)
```

```python
rse = metrics.r2_score(Y_test, test_data_prediction)
mae = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error: ",rse)
print("Mean absolute error: ",mae)
```
